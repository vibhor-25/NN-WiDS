```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/fruit_production.csv')
df["Total"]=df['Mangoes_ton']+df['Oranges_ton']
df
```

```python
x1=np.array(df['Temperature_F']/100)
x2=np.array(df['Rainfall_mm']/100)
x3=np.array(df['Humidity_percent']/100)

y1=np.array(df['Mangoes_ton']/100)
y2=np.array(df['Oranges_ton']/100)
y3=np.array(df['Total']/100)

x=np.array([x1,x2,x3])
y=np.array([y1,y2,y3])

n=len(x1)
```

```python
a=np.array([0.0,0.0,0.0])
b=0.0
d=np.array([0.0]*3)

def j1(x,y1,a,b,d,n):
  return np.sum((np.dot(a,x)+b-y1)**2)/(n*2)

def dj1(x,y1,a,b,d,n):
  for i in range(3):
    d[i]=np.sum((np.dot(a,x)+b-y1)*x[i])/n
  return

def djb1(x,y1,a,b,d,n):
  return np.sum((np.dot(a,x)+b-y1))/n

lr = 0.01
dj1(x,y1,a,b,d,n)

while(not(abs(djb1(x,y1,a,b,d,n))<10**(-3) and max(abs(d))<10**(-3))):
  dj1(x,y1,a,b,d,n)
  tempb = b - lr*djb1(x,y1,a,b,d,n)
  a = a - lr * d
  b = tempb

print(a,b,j1(x,y1,a,b,d,n))

```
