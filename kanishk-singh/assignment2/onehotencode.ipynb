```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('/content/house_rent_dataset.csv')
df = df.drop(['Posted On','Floor','Area Locality','Tenant Preferred','Point of Contact'],axis = 1)

df
```

```python
print(df['Area Type'].unique(),'\n',df['City'].unique(),'\n',df['Furnishing Status'].unique())
```

```python
df[['SA','CA','BA','K','M','B','D','C','H']]=0
df.loc[df['Furnishing Status']=='Unfurnished','Furnishing Status']=0
df.loc[df['Furnishing Status']=='Semi-Furnished','Furnishing Status']=1
df.loc[df['Furnishing Status']=='Furnished','Furnishing Status']=2

df.loc[df['City']=='Kolkata','K']=1
df.loc[df['City']=='Mumbai','M']=1
df.loc[df['City']=='Bangalore','B']=1
df.loc[df['City']=='Delhi','D']=1
df.loc[df['City']=='Chennai','C']=1
df.loc[df['City']=='Hyderabad','H']=1

df.loc[df['Area Type']=='Super Area','SA']=1
df.loc[df['Area Type']=='Carpet Area','CA']=1
df.loc[df['Area Type']=='Built Area','BA']=1

df = df.drop(['Area Type','City'],axis = 1)
df = df.rename({"Furnishing Status":"FS"},axis=1)
df
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df[['BHK', 'Size', 'FS', 'Bathroom', 'SA', 'CA', 'BA', 'K', 'M',
       'B', 'D', 'C', 'H']],df['Rent'],test_size=0.2,random_state=69)

xtest['Size'] = xtest['Size']/1000
ytest = ytest/10000

xtest , ytest
```

```python
columns = ['BHK', 'Size', 'FS', 'Bathroom', 'SA', 'CA', 'BA', 'K', 'M',
       'B', 'D', 'C', 'H']

x = np.vstack([*[xtrain[col].values for col in columns]])
y = np.array(*[ytrain.values])

x[1] = x[1]/1000
y = y/10000

n = len(x[0])
x,y
```

```python
co = np.array([0.0]*13)
dco = np.array([0.0]*13)
c = 0

def j(x,y,co,c,n):
  return np.sum((np.dot(co,x)+c-y)**2)/(2*n)

def dj(x,y,co,c,n):
  for i in range(13):
    dco[i] = np.sum((np.dot(co,x)+c-y)*x[i])/n
  return

def dc(x,y,co,c,n):
  return np.sum((np.dot(co,x)+c-y))/n

j(x,y,co,c,n)
```

```python
dj(x,y,co,c,n)
lr = 0.1

while not(max(abs(dco)<10**(-3)) and abs(dc(x,y,co,c,n))<10**(-3)):
  dj(x,y,co,c,n)

  tc = c - lr*dc(x,y,co,c,n)
  co = co - lr*dco
  c = tc

  print(j(x,y,co,c,n))

print(j(x,y,co,c,n),co,c)
```
