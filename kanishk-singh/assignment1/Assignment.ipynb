### Q1 - FizzBuzz

```python
for i in range(1,101):
  if i%15==0:
    print("FizzBuzz")
  elif i%3==0:
    print("Fizz")
  elif i%5==0:
    print("Buzz")
  else :
    print(i)
```

### Q2- Matrix Operations

```python
import numpy as np

a = np.random.randint(1,11,(3,3))
b = np.random.randint(1,11,(3,3))

print(a,'\n'*2,b,'\n'*2,a+b,'\n'*2,a*b,'\n')

dot = np.linalg.matmul(a,b)
print(dot)

print('\n',int(np.linalg.det(a)))
```

### Q4 - Plotting a curve

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.01)
y = x**3+2*(x**2)+3*x+2

xtick = np.arange(-5,5,1)
ytick = np.arange(-100,201,25)

plt.figure(1)
plt.plot(x,y)

plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black')

plt.xticks(xtick)
plt.yticks(ytick)

plt.show()

```

```python
x = np.arange(-5,5,0.01)
y= np.tan(x)/x
xtick = np.arange(-5,6,1)
ytick = np.arange(-10,11,2)

plt.figure(2)
plt.plot(x,y)

plt.xlim(-5,5)
plt.ylim(-10,10)

plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black')

plt.xticks(xtick)
plt.yticks(ytick)

plt.show()
```

```python
y= np.arange(-5,5,0.01)
x = 3*y + np.sin(y)
xtick = np.arange(-15,16,5)
ytick = np.arange(-5,6,1)

plt.figure(3)
plt.plot(x,y)

plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black')

plt.xticks(xtick)
plt.yticks(ytick)

plt.show()
```
