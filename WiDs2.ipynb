```python
df=pd.read_csv('log_reg.csv')
```

```python
df
```

```python
plt.scatter(df[df['color']==0]['x'], df[df['color']==0]['y'], color='red', label='Class 0')
plt.scatter(df[df['color']==1]['x'], df[df['color']==1]['y'], color='blue', label='Class 1')
```

```python
x=df[['x','y']]
y=df['color']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
predictions=model.predict(x_test)
```

```python
accuracy_score(y_test,predictions)
```
