```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
data = pd.read_csv('fruit_production.csv')

data.head()
data.info()
data.describe()



```

```python
X = data[["Temperature_F", "Rainfall_mm", "Humidity_percent"]]
y = data["Mangoes_ton"]
```

```python
x = data[["Temperature_F", "Rainfall_mm", "Humidity_percent"]]
Y = data["Oranges_ton"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=40
)
```

```python
x_train, x_test, Y_train, Y_test = train_test_split(
    x, Y, test_size=0.2, random_state=44
)
```

```python
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
```

```python
model_2 = LinearRegression()
model_2.fit(x_train, Y_train)
```

```python
print("Coefficients:", model_1.coef_)
print("Intercept:", model_1.intercept_)
```

```python
print("Coefficients:", model_2.coef_)
print("Intercept:", model_2.intercept_)
```

```python
print("Train R²:", model_1.score(X_train, y_train))
print("Test R²:", model_1.score(X_test, y_test))
y_pred = model_1.predict(X_test)
print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

```python
print("Train R²:", model_2.score(x_train, Y_train))
print("Test R²:", model_2.score(x_test, Y_test))
Y_pred = model_2.predict(x_test)
print("Predictions:", Y_pred)
print("Actual:", Y_test.values)
```

```python
sample = pd.DataFrame({
    'Temperature_F': [5],
    'Rainfall_mm': [14],
    'Humidity_percent': [2]
})

predicted_production_1 = model_1.predict(sample)
predicted_production_2 = model_2.predict(sample)
print("Predicted mango production :", predicted_production_1[0])
print("Predicted orange production:", predicted_production_2[0])
```
