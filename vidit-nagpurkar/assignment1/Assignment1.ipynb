```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

Exercise 1: FizzBuzz

```python
for i in range(100):
    if (i + 1)% 15 == 0:
        print("FizzBuzz")
    elif (i + 1) % 3 == 0:
        print("Fizz")
    elif (i + 1) % 5 == 0:
        print("Buzz")
    else:
        print(i+1)
```

Exercise 2: Matrix Operations

```python
MatrixA = np.random.randint(1, 11, (3, 3))
MatrixB = np.random.randint(1, 11,(3, 3))

print("Addition:")
print(MatrixA + MatrixB)

print("Element-wise multiplication:")
print(MatrixA *MatrixB)

print("Dot product:")
print(MatrixA@ MatrixB)

print("Determinant of matrix A:")
print(np.round(np.linalg.det(MatrixA)))
```

Exercise 3: Pandas Manipulation

```python
csv = pd.read_csv("https://drive.google.com/uc?id=1OekIEbM6phQNRAZZOfh4optv5Cjh_iQG")

numbertype = []
for value in csv[csv.columns[0]]:
    if value < 0.3:
        numbertype.append("Low")
    elif value < 0.7:
        numbertype.append("Medium")
    else:
        numbertype.append("High")

csv["type"] = numbertype
csv_deleted = []
for i in range(len(csv)):
    if numbertype[i]!= "Low":
        csv_deleted.append(csv[csv.columns[0][i]])

csv_manipulated = pd.DataFrame(csv_deleted)
csv_manipulated.to_csv("final.csv", index=False)
files.download("final.csv")
```

Exercise 4: Plotting a curve

```python
x = np.linspace(-5,5,100)
y1 = x**3 + 2*x**2 + 3 * x + 2
y2 = np.tan(x)/x
y3 = np.linspace(-np.pi,np.pi,100)
x3 = np.sin(y3) + 3*y3

plt.subplot(1,3,1)
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.subplot(1,3,3)
plt.plot(x3,y3)

plt.show()
```
