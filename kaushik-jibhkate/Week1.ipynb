```python
for i in range(1,101,1):
  if i%3==0 and i%5==0:
    print("FizzBuzz")
  elif i%3==0:
    print("Fizz")
  elif i%5==0:
    print("Buzz")
  else:
    print(i)
```

```python
import numpy as np
A = np.random.randint(0,10,size=(3,3))
B = np.random.randint(0,10,size=(3,3))
print(A)
print(B)
sum = A+B
print(sum)
element_wise_multiplication = A*B
print(element_wise_multiplication)
dot_product = np.dot(A,B)
print(dot_product)
print(np.linalg.det(A))
```

```python
import pandas as pd
df = pd.read_csv('data_assg1.csv')
df.loc[df['value']<=0.3, 'degree'] = 'Low'
df.loc[(df['value']>0.3) & (df['value']<=0.7), 'degree'] = 'Medium'
df.loc[df['value']>0.7, 'degree'] = 'High'
print(df)

```

```python
df.drop(df[df['degree']=='Low'].index)
print(df)
df.to_csv('data_assg1_new.csv',index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [i for i in range(1,100,1)]
y = [i*i*i+2*i*i+3*i+2 for i in range(1,100,1)]
plt.figure(1)
plt.plot(x,y)

x2 = [i for i in range(1,100,1)]
y2 = [np.tan(i)/i for i in range(1,100,1)]
plt.figure(2)
plt.plot(x2,y2)

x3 = [np.sin(i)+3*i for i in range(1,10)]
y3 = [i for i in range(1,10)]
plt.figure(3)
plt.plot(x3,y3)
plt.show()
```
